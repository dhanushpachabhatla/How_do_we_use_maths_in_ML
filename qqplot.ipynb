#How to say if a column follow normal distribution??

* can draw a distribution graph or can use qqplot from lib or can draw qqplot using scatterplot as below

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
```

```python
sns.kdeplot(df['sepal_length'])
```

```python
temp = sorted(df['sepal_length'].tolist())
```

calculating percentiles

```python
y_quant = []
for i in range(1,101):
  y_quant.append(np.percentile(temp,i))
```

```python
samples = np.random.normal(loc=0,scale = 1, size = 1000)
#loc = mean
#scale = standart deviation
```

```python
x_quant = []
for i in range(1,101):
  x_quant.append(np.percentile(samples,i))
```

```python
sns.scatterplot(x = x_quant, y = y_quant)
```

* if it follows a diagonal straight line then yes it is normal but here it is close to diagonal line but not perrfect so it is type of normal diistribution but not a perfect normal distribution

* here we are comparing our column with a normaly distributed data

#or can use lib for qqplot

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt

fig = sm.qqplot(df['sepal_length'], line='45', fit=True)
#read qqplot documentation for more parameters info

plt.title('QQ PLOT')
plt.xlabel('Theoretical Qunatities')
plt.ylabel('Sample Quantities')

plt.show()
```
