#Normal distribution use in DS

uses :
* outlier detection
* assumptions on data for ML-algorithms -> Linear Regression and GMM
* Hypothesis Testing
* Central Limit Theorem

using titanic dataset

```python
 tf = sns.load_dataset("titanic")
```

```python
tf.head()
```

```python
tf['age'].skew()
```

* basically we use ND to find outliers in a particular columns

```python
tf['age'].mean() + 3*tf['age'].std()
```

```python
tf['age'].mean() - 3*tf['age'].std()
```

from the above we can say 99.7 observations have age <= 73 and >= -13 so we can assume the other part of population as outliers and can remove them
